```python
from tkinter import *
from tkinter.messagebox import showinfo

def get_data():
    try:
        # Retrieve and convert each DoubleVar to a float value
        name = st_name_1.get()
        urdu = float(urdu_name_1.get())
        english = float(english_name_1.get())
        science = float(science_name_1.get())
        computer = float(computer_name_1.get())
        math = float(math_name_1.get())

        # Calculate total and percentage
        total = urdu + english + science + computer + math
        percentage = (total / 500) * 100

        # Determine division
        if percentage >= 60:
            division = "1st Division"
        elif percentage >= 50:
            division = "2nd Division"
        elif percentage >= 40:
            division = "3rd Division"
        else:
            division = "Fail"

        # Display the result
        messagebox(name, total, percentage, division)

    except ValueError:
        showinfo(title="Error", message="Please enter valid numbers for all subject scores.")

def messagebox(name, total, percentage, division):
    report = f"""
    Name: {name}
    Total: {total}
    Percentage: {percentage:.2f}%
    Division: {division}
    """
    showinfo(title="Report Card", message=report)

# Tkinter window setup
win = Tk()
win.title("College Report Card")
win.config(bg='lightblue')
win.geometry("600x600")
win.resizable(False, False)

# Title
college_name = Label(win, text="Science College", font=("Ariel", 40, "bold"), bg='lightblue')
college_name.place(x=100, y=20, height=60, width=400)

# Student name entry
st_name = Label(win, text="Student Name", font=("Ariel", 20, "bold"))
st_name.place(x=10, y=100, height=40, width=200)
st_name_1 = StringVar()
st_name_entry = Entry(win, font=("Ariel", 20, "bold"), textvariable=st_name_1)
st_name_entry.place(x=230, y=100, height=40, width=300)

# Subject scores
subject_labels = ["Urdu", "English", "Science", "Computer", "Math"]
subject_vars = [DoubleVar(), DoubleVar(), DoubleVar(), DoubleVar(), DoubleVar()]
y_positions = [240, 300, 360, 420, 480]

for i, subject in enumerate(subject_labels):
    label = Label(win, text=subject, font=("Ariel", 20, "bold"))
    label.place(x=10, y=y_positions[i], height=40, width=200)

    entry = Entry(win, font=("Ariel", 20, "bold"), textvariable=subject_vars[i])
    entry.place(x=230, y=y_positions[i], height=40, width=300)

urdu_name_1, english_name_1, science_name_1, computer_name_1, math_name_1 = subject_vars

# Done button
button = Button(win, text="Done", font=("Ariel", 20, 'bold'), command=get_data)
button.place(x=200, y=540, height=40, width=100)

win.mainloop()
```
