```python
# import necessary libraries
import tkinter as tk
from PIL import ImageTk,Image
from tkinter import messagebox
from sklearn.datasets import load_iris
from sklearn.tree import  DecisionTreeClassifier
import joblib
iris=load_iris()

X=iris.data
y=iris.target

dt=DecisionTreeClassifier()
dt.fit(X,y)
# save the model
joblib.dump(dt,"iris_model.joblib")

# Create main window
root = tk.Tk()
root.title("Iris Flower Predictor")  # title of screen
root.geometry("500x230")  # size of screen

# background color of window
root.config(bg='lightblue')
# Header for window
header=tk.Label(root,text="Iris Flower Predictor",bg='lightblue',fg='black',font=("Ariel",15,'bold'))
header.pack()

# create frame to make gui
frame=tk.Frame(root,bg="lightblue")
frame.pack()


#make gui
label1=tk.Label(frame,text="Sepal Length",bg='lightblue',fg='black',font=("Ariel",15,'bold'))
label1.grid(row=0,column=0,pady=10)

entry1=tk.Entry(frame,width=10,bg='grey',fg='white',font=("Ariel",15,'bold'))
entry1.grid(row=0,column=1,pady=10)


 # ***************************
label2=tk.Label(frame,text="Sepal Width",bg='lightblue',fg='black',font=("Ariel",15,'bold'))
label2.grid(row=0,column=2,pady=10)

entry2=tk.Entry(frame,width=10,bg='grey',fg='white',font=("Ariel",15,'bold'))
entry2.grid(row=0,column=3,pady=10)


#*******************************
label3=tk.Label(frame,text="Petal Length",bg='lightblue',fg='black',font=("Ariel",15,'bold'))
label3.grid(row=1,column=0,pady=10)

entry3=tk.Entry(frame,width=10,bg='grey',fg='white',font=("Ariel",15,'bold'))
entry3.grid(row=1,column=1,pady=10)

# *********************************

label4=tk.Label(frame,text="Petal Width",bg='lightblue',fg='black',font=("Ariel",15,'bold'))
label4.grid(row=1,column=2,pady=10)

entry4=tk.Entry(frame,width=10,bg='grey',fg='white',font=("Ariel",15,'bold'))
entry4.grid(row=1,column=3,pady=10)

# label for result
result=tk.Label(root,text="Predicted Species:",bg='lightblue',fg='black',font=("Ariel",15,'bold'))
result.pack(pady=10)



# load the saved model
model=joblib.load("iris_model.joblib")

# Create function for predict species
def Predict_species():
    sepal_length = int(entry1.get())
    sepal_width = int(entry2.get())
    petal_length = int(entry3.get())
    petal_width = int(entry4.get())
    prediction= model.predict([[sepal_length,sepal_width,petal_length,petal_width]])
    result.config(text="Predicted Species:"+iris.target_names[prediction[0]])



# create button
button=tk.Button(text="Predict",bg='blue',activebackground='lightblue',borderwidth=3,font=("Ariel",15,'bold'),command=Predict_species)
button.pack()





root.mainloop()
```
