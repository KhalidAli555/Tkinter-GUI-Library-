```python
# import necessary libraries
import tkinter as tk
from PIL import ImageTk,Image
from tkinter import messagebox
from sklearn.datasets import load_iris
from sklearn.tree import  DecisionTreeClassifier
import joblib
iris=load_iris()

X=iris.data
y=iris.target

dt=DecisionTreeClassifier()
dt.fit(X,y)
# save the model
joblib.dump(dt,"iris_model.joblib")

# Create main window
root = tk.Tk()
root.title("Iris Flower Predictor")  # title of screen
root.geometry("300x500")  # size of screen

# minimium size of screen
root.minsize(300,300)

# maximium size of screen
root.maxsize(600,600)

# change background color of window
root.configure(bg='lightblue')

# ///////////////////////////////////////

# Enter Sepal Length Label
label1=tk.Label(root, text="Sepal Length", bg="lightblue", fg="black", font=("Arial", 12, "bold"))
label1.pack(pady=3)

# sepal length Entry widget (Don't call .pack() here)
entry1 = tk.Entry(root, width=10, bg='grey', fg='white', font=("Arial", 15, "bold"))
entry1.pack(pady=3)
# **********************************************************************
# Enter Sepal width Label
label2=tk.Label(root, text="Sepal Width", bg="lightblue", fg="black", font=("Arial", 12, "bold"))
label2.pack(pady=3)

# sepal width Entry widget (Don't call .pack() here)
entry2 = tk.Entry(root, width=10, bg='grey', fg='white', font=("Arial", 15, "bold"))
entry2.pack(pady=3)

# >>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>

# Enter Petal Length Label
label3=tk.Label(root, text="Patel Length", bg="lightblue", fg="black", font=("Arial", 12, "bold"))
label3.pack(pady=3)

# sepal length Entry widget (Don't call .pack() here)
entry3 = tk.Entry(root, width=10, bg='grey', fg='white', font=("Arial", 15, "bold"))
entry3.pack(pady=3)

 # """"""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""
# Enter petal width Label
label4=tk.Label(root, text="Patel Width", bg="lightblue", fg="black", font=("Arial", 12, "bold"))
label4.pack(pady=3)

# petal width Entry widget (Don't call .pack() here)
entry4 = tk.Entry(root, width=10, bg='grey', fg='white', font=("Arial", 15, "bold"))
entry4.pack(pady=3)

# ***********************************

# label for result
species=tk.Label(root, text="", bg="lightblue", fg="black", font=("Arial", 12, "bold"))
species.pack(pady=3)



# load the saved model
model=joblib.load("iris_model.joblib")

# **************************************************

# Create function for predict species
def Predict_species():
    sepal_length = int(entry1.get())
    sepal_width = int(entry2.get())
    petal_length = int(entry3.get())
    petal_width = int(entry4.get())
    prediction= model.predict([[sepal_length,sepal_width,petal_length,petal_width]])
    species.config(text="Predicted Species:"+iris.target_names[prediction[0]])
# Create predict Button
predict_button = tk.Button(root, text="Predict",bg="lightgreen", command=Predict_species)
predict_button.pack(pady=10)



root.mainloop()
```
