```python
from tkinter import *

# Function to update the input field
def button_click(number):
    current = entry.get()
    entry.delete(0, END)
    entry.insert(0, current + str(number))

# Function to clear the input field
def button_clear():
    entry.delete(0, END)

# Function to perform the calculation
def button_equal():
    try:
        result = eval(entry.get())  # Evaluate the expression
        entry.delete(0, END)
        entry.insert(0, str(result))
    except:
        entry.delete(0, END)
        entry.insert(0, "Error")

# Function to handle operations
def button_operation(op):
    current = entry.get()
    entry.delete(0, END)
    entry.insert(0, current + op)

# Initialize the main window
win = Tk()
win.title("Calculator")
win.geometry("400x500")
win.resizable(False, False)

# Entry field
entry = Entry(win, width=16, font=("Arial", 24), borderwidth=5, relief="ridge", justify='right')
entry.grid(row=0, column=0, columnspan=4, pady=20)

# Number buttons
buttons = [
    ('7', 1, 0), ('8', 1, 1), ('9', 1, 2),
    ('4', 2, 0), ('5', 2, 1), ('6', 2, 2),
    ('1', 3, 0), ('2', 3, 1), ('3', 3, 2),
    ('0', 4, 1)
]

for (text, row, col) in buttons:
    Button(win, text=text, font=("Arial", 18), command=lambda t=text: button_click(t)).grid(row=row, column=col, sticky="nsew")

# Operation buttons
Button(win, text="+", font=("Arial", 18), command=lambda: button_operation("+")).grid(row=1, column=3, sticky="nsew")
Button(win, text="-", font=("Arial", 18), command=lambda: button_operation("-")).grid(row=2, column=3, sticky="nsew")
Button(win, text="*", font=("Arial", 18), command=lambda: button_operation("*")).grid(row=3, column=3, sticky="nsew")
Button(win, text="/", font=("Arial", 18), command=lambda: button_operation("/")).grid(row=4, column=3, sticky="nsew")

# Equal button
Button(win, text="=", font=("Arial", 18), command=button_equal).grid(row=4, column=2, sticky="nsew")

# Clear button
Button(win, text="C", font=("Arial", 18), command=button_clear).grid(row=4, column=0, sticky="nsew")

# Make buttons expandable
for i in range(5):
    win.grid_rowconfigure(i, weight=1)
    win.grid_columnconfigure(i, weight=1)

win.mainloop()
```
